```python
def fact(n):
    if n==0:
        return 1
    return n* fact(n-1)
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

```

```python
n=int(input("Enter a number: "))
print(fact(n))
for i in range(7):
    print(fib(i), end=" ")
```

```python
def pal(n):
    x=n[::-1]
    if x==n:
        return 1
    else:
        return 0
```

```python
n = input("Enter: ")

if pal(n) == 1:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n-1)
```

```python
x=int(input("Enter the power: "))
n=int(input("Enter a number: "))
print(power(x, n))
```

```python
def sumL(L):
    return sum(L)
```

```python
def Max(L):
    return max(L)
```

```python
n=int(input("Enter the number of elements: "))
L = []
for i in range(n):
    i=int(input("Enter a number: "))
    L.append(i)
print(sumL(L))
print(Max(L))
```

```python
def srev(s):
    x = s[::-1]
    return x
```

```python
s=input("Enter a string: ")
print(srev(s))
```
