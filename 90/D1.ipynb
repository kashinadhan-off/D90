```python
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter operation \n1:+ \n2:- \n3:* \n4://\n"))

if z == 1:
    print(x + y)
elif z == 2:
    print(x - y)
elif z == 3:
    print(x * y)
elif z == 4:
    if y != 0:
        print(x / y)
```

```python
x = float(input("Enter the deposit amount: "))
y = float(input("Enter the interest rate: "))
z = int(input("Enter number of years: "))

si = (x * y * z) / 100

print("Simple interest =", si)
```

```python
x = int(input("Enter the first number: "))
y = int(input("Enter the second number: "))
z = int(input("Enter the third number: "))

if x > y and x > z:
    print("Largest number,x =", x)
elif y > x and y > z:
    print("Largest number,y =", y)
else:
    print("Largest number,z =", z)
```
