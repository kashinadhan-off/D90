```python
import json
```

```python
def load_tasks():
    try:
        with open("D6tasks.txt", "r") as f:
            return json.load(f)
    except:
        return {}
```

```python
def save_tasks(tasks):
    with open("D6tasks.txt", "w") as f:
        json.dump(tasks, f, indent=4)
```

```python
def add_task():
    tasks = load_tasks()

    t = input("Enter task: ")
    tasks[t] = False

    save_tasks(tasks)
```

```python
def view_tasks():
    tasks = load_tasks()

    for i, task in enumerate(tasks):
        status = "✔" if tasks[task] else "❌"
        print(i+1, task, status)
```

```python
def update_task():
    tasks = load_tasks()

    t = input("Enter task name: ")

    if t in tasks:
        tasks[t] = not tasks[t]   # toggle True ↔ False
        save_tasks(tasks)
        print("Updated")
    else:
        print("Task not found")
```

```python
def delete_task():
    tasks = load_tasks()

    t = input("Enter task name: ")

    if t in tasks:
        del tasks[t]
        save_tasks(tasks)
        print("Deleted")
    else:
        print("Not found")
```

```python

    while True:
        print("-----------ENTER CHOICE-----------")
        print("1. Add\n2. View\n3. Delete\n4. Update\n5. Exit")
        print("---------TO-DO LIST---------------")
        try:
            ch = int(input("Enter choice: "))
        except:
            print("Invalid input")
            continue

        match ch:
            case 1:
                add_task()
            case 2:
                view_tasks()
            case 3:
                delete_task()
            case 4:
                update_task()
            case 5:
                print("Exiting...")
                break
            case _:
                print("Invalid choice")
```
